```python
# 1.Move info in the separate files into their corresponding tabs of this "Consolidation" file
import pandas as pd

# Read the data from the source Excel file
source_file = 'Finance.xlsx'
source_sheet = 'Sheet1'
df = pd.read_excel(source_file, sheet_name=source_sheet)

# Specify the destination file and sheet name
destination_file = 'Consolidation.xlsx'
destination_sheet = 'Finance_consolidation'

with pd.ExcelWriter(destination_file, engine='openpyxl', mode='a',if_sheet_exists='replace') as writer:
    # Write the DataFrame to the destination sheet
    df.to_excel(writer, sheet_name=destination_sheet, index=False)

# Read the data from the source Excel file
source_file = 'IT.xlsx'
source_sheet = 'Sheet1'
df = pd.read_excel(source_file, sheet_name=source_sheet)

# Specify the destination file and sheet name
destination_file = 'Consolidation.xlsx'
destination_sheet = 'IT_consolidation'

with pd.ExcelWriter(destination_file, engine='openpyxl', mode='a',if_sheet_exists='replace') as writer:
    # Write the DataFrame to the destination sheet
    df.to_excel(writer, sheet_name=destination_sheet, index=False)
# Read the data from the source Excel file
source_file = 'Legal.xlsx'
source_sheet = 'Sheet1'
df = pd.read_excel(source_file, sheet_name=source_sheet)

# Specify the destination file and sheet name
destination_file = 'Consolidation.xlsx'
destination_sheet = 'Legal_consolidation'

with pd.ExcelWriter(destination_file, engine='openpyxl', mode='a',if_sheet_exists='replace') as writer:
    # Write the DataFrame to the destination sheet
    df.to_excel(writer, sheet_name=destination_sheet, index=True)

# 2.Consolidate the details of tabs in the individual tabs and paste into "Database" tab as shown in the tab. I name it Database 2

import pandas as pd

def combine_tabs(excel_file, tabs_to_combine, target_tab):


    # Read the Excel file
    with pd.ExcelWriter(excel_file, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
        combined_data = pd.DataFrame()

        # Combine data from specified tabs
        for tab_name in tabs_to_combine:
            df = pd.read_excel(excel_file, sheet_name=tab_name)
            df['Cost Center'] = tab_name.split("_")[0]
            df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
            combined_data = pd.concat([combined_data, df])

        # Write the combined data to the target tab
        combined_data.to_excel(writer, sheet_name=target_tab, index=False)

if __name__ == "__main__":
    excel_file = "Consolidation.xlsx"
    tabs_to_combine = ["IT_consolidation", "Finance_consolidation","Legal_consolidation"]
    target_tab = "Database2"

    combine_tabs(excel_file, tabs_to_combine, target_tab)

#3.Create a pivot table for vendor as shown in "Pivot Table Vendor" tab. Please use Google Colab or Google Apps script
import pandas as pd
data = pd.read_excel('Consolidation.xlsx', sheet_name='Database')
df = pd.DataFrame(data)


# Create the pivot table
df = pd.pivot_table(df, index=['Vendor'], aggfunc='sum', sort=False)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
# Calculate the sum of each column
sum_row = df.sum()

# Append the sum row to the DataFrame
df.loc['Grand Total'] = sum_row.drop("Account Name")

# another way df = df.groupby('Vendor').sum()


#print(df)
# 4.Sum up the amounts in IT, Finance and Legal tabs in "P&L" tab as shown in the tab.

import pandas as pd
data = pd.read_excel('Consolidation.xlsx', sheet_name='Database')
df = pd.DataFrame(data)
df = df.drop('Vendor', axis=1)


# Create the pivot table
df = pd.pivot_table(df, index=['Account Name'], aggfunc='sum', sort=False)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

print(df)




```
